```python
# your code here
customerlabels = ['Cheap', 'Regular', 'Preferred', 'VIP']
quantiles = [.25,.50,.75,.95, 1]
binning = pd.qcut(orders["amount_spent"],quantiles, labels= customerlabels)
orders['customer_type'] = binning
orders[['amount_spent','customer_type']].head(10)
```

```python
orders.groupby(["customer_type"])["amount_spent"].sum()
```

## Q2: How to identify which country has the most VIP Customers?

```python
countries = orders.groupby(['Country','customer_type'])['CustomerID'].nunique().reset_index()
countries
```

```python
VIP = countries[countries['customer_type']=='VIP']
VIP
```

```python
VIP_ROW = VIP[VIP['CustomerID']==VIP['CustomerID'].max(axis=0, skipna=True)]
VIP_ROW
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
# We already have determined the Country with the most VIP, now we have to go for the preferred and concat them: 
PREF = countries[countries['customer_type']=='Preferred']
PREF
```

```python
total = pd.concat([VIP, PREF], axis=0, sort=False)
total
```

```python
VIP_and_PREF = total.groupby(['Country']).agg(customer_type_count=('CustomerID','sum')).reset_index()
VIP_and_PREF
```

```python
VIP_and_PREF_max = VIP_and_PREF[VIP_and_PREF['customer_type_count']==VIP_and_PREF['customer_type_count'].max(axis=0, skipna=True)]
VIP_and_PREF_max
```
